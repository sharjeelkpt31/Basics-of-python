<h1>2-Python Data Types</h1>
<p>Python Data types are the classification or categorization of data items. It represents the kind of value that tells what operations can be performed on a particular data.</p>

<b>Numeric – int, float, complex</b>

<b>Sequence Type – string, list, tuple</b>

<b>Mapping Type – dict</b>

<b>Boolean – bool</b>

<b>Set Type – set, frozenset</b>

<b>Binary Types – bytes, bytearray, memoryview</b

```python
# Numeric  int,float,complex
a = 4 # integer
b = 4.5 # float
c = 4j # complex number 

print(type(a))
print(type(b))
print(type(c))
```

```python
#Sequence Type  
s = "GfG"

print(s[1]) # access 2nd char
s1 = s + s[0] # update
print(s1) # print
```

```python
# Creating a String
s1 = 'GfG'
s2 = "GfG"
print(s1)
print(s2)
```

```python
# Multi-line Strings
s = """I am Learning
Python String on GeeksforGeeks"""
print(s)

s = '''I'm a 
sharjeel'''
print(s)
```

```python
#Python Lists
# Creating a Python list with different data types
a = [10, 20, "GfG", 40, True]

print(a)

# Accessing elements using indexing
print(a[0])  # 10
print(a[1])  # 20
print(a[2])  # "GfG"
print(a[3])  # 40
print(a[4])  # True

# Checking types of elements
print(type(a[2]))  # str
print(type(a[4]))  # bool
```

```python
# Python Tuples
# Creating an empty Tuple
tup = ()
print(tup)

# Using String
tup = ('Geeks', 'For')
print(tup)

# Using List
li = [1, 2, 4, 5, 6]
print(tuple(li))

# Using Built-in Function
tup = tuple('Geeks')
print(tup)
```

```python
# Mapping Type Dict
d = {1: 'do good', 2: 'have', 3: 'good'}
print(d)
```

```python
# How to Create a Dictionary
# create dictionary using { }
d1 = {1: 'A', 2: 'B', 3: 'C'}
print(d1)

# create dictionary using dict() constructor
d2 = dict(a = "G", b = "for", c = "N")
print(d2)
```

```python
# Boolean – bool
a = True
print(type(a))

b = False
print(type(b))
```

```python
# Set Type – set, frozenset
set1 = {1, 2, 3, 4}
print(set1)
```

```python
set1 = set("SALMAN")
print(set1)

# Creating a Set with the use of a List
set1 = set(["HI", "HELLO", "HI"])
print(set1)

# Creating a Set with the use of a tuple
tup = ("HI", "HELLO", "HI")
print(set(tup))

# Creating a Set with the use of a dictionary
d = {"HEN": 1, "GOAT": 2, "COW": 3}
print(set(d))
```

```python
# frozenset() in Python
animals = frozenset(["cat", "dog", "lion"])
print("cat" in animals) 
print("elephant" in animals)  
```

```python
# Binary Types – bytes, bytearray, memoryview
a = "CHERRY"

 # UTF-8 encoding is used
b = bytes(a, 'utf-8') 
print(b)
```

```python
# bytearray() function – Python
# Create a bytearray from a string
ba = bytearray("best wishes!", "utf-8")
print(ba)  # Output: bytearray(b'geeks for geeks!')

# Modify the bytearray 
ba[1] = 105  # 'i' in ASCII is 105
print(ba)  
```

```python
#cmemoryview() in Python
# Creating a memory view of a bytearray
data = bytearray("Hello", "utf-8")   
mv = memoryview(data)                

print(mv[0])           
mv[1] = 105           
print(data)     
```

```python
import array

# Creating an array of integers
arr = array.array('i', [1, 2, 3, 4])   
mv=memoryview(arr)

# Reinterpreting the data as unsigned bytes
mv_cast = mv.cast('B')                
print(mv_cast.tolist()) 
```
