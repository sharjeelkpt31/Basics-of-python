<h1>Basic Arithmetic Operations</h1>
<p>1- Arithmetic Operators :</p>
Python operators are fundamental for performing mathematical calculations. Arithmetic operators are symbols used to perform mathematical operations on numerical values. Arithmetic operators include addition (+), subtraction (-), multiplication (*), division (/), and modulus (%).

```python
#Addition Operator
val1 = 2
val2 = 3

# using the addition operator
res = val1 + val2
print(res)
```

```python
#Subtraction Operator
val1 = 8
val2 = 3

# using the subtraction operator
res = val1 - val2
print(res)
```

```python
#Multiplication Operator
val1 = 2
val2 = 3

# using the multiplication operator
res = val1 * val2
print(res)
```

```python
#Division Operator 

print(5/5)
print(10/2)
print(-10/2)
print(20.0/2)
```

```python
# Exponentiation Operator
val1 = 2
val2 = 3

# using the exponentiation operator
res = val1 ** val2
print(res)
```
