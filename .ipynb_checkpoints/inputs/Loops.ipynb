<h1>Loops in Python</h1>

<h1>While Loop in Python</h1>
<p>Python While Loop is used to execute a block of statements repeatedly until a given condition is satisfied. When the condition becomes false, the line immediately after the loop in the program is executed.</p>


```python
# Python example for while loop
count = 0
while (count < 3):
    count = count + 1
    print("Hello friend")
```

<p><b>Infinite while Loop in Python</p>

```python
age = 28

# the test condition is always True
while age > 19:
    print('Infinite Loop')
```

<h1>while loop with continue statement</h1>

```python
# Prints all letters except 'e' and 's'
i = 0
a = 'friend'

while i < len(a):
    if a[i] == 'e' or a[i] == 's':
        i += 1
        continue
        
    print(a[i])
    i += 1
```

<h1>while loop with break statement</h1>

```python
# break the loop as soon it sees 'e'
# or 's'
i = 0
a = 'friend'

while i < len(a):
    if a[i] == 'e' or a[i] == 's':
        i += 1
        break
        
    print(a[i])
    i += 1
```

<h1>while loop with pass statement</h1>

```python
# An empty loop
a = 'hi everyone'
i = 0

while i < len(a):
    i += 1
    pass
  
print('Value of i :', i)
```

<h1>while loop with else</h1>

```python
# Python program to demonstrate
# while-else loop

i = 0
while i < 4:
    i += 1
    print(i)
else:  # Executed because no break in for
    print("No Break\n")

i = 0
while i < 4:
    i += 1
    print(i)
    break
else:  # Not executed as there is a break
    print("No Break")
```

<h1>For Loop in Python</h1>
<p>Python For Loops are used for iterating over a sequence like lists, tuples, strings, and ranges.

<br>For loop allows you to apply the same operation to every item within loop.
Using For Loop avoid the need of manually managing the index.
For loop can iterate over any iterable object, such as dictionary, list or any custom iterators.</br>
For Loop Example:</p>

```python
s = ["hello", "hi", "hello"]

# using for loop with string
for i in s:
    print(i)
```

<h1>Python For Loop with String</h1>

```python
s = "forever"
for i in s:
    print(i)
```

<h1>Using range() with For Loop</h1>

```python
for i in range(0, 10, 2):
    print(i)
```

<h1>Continue with For Loop

```python
# Prints all letters except 'e' and 's'

for i in 'foreveris':

    if i == 'e' or i == 's':
        continue
    print(i)
```

<h1>Break with For Loop

```python
for i in 'salman':

    # break the loop as soon it sees 'e'
    # or 's'
    if i == 'e' or i == 's':
        break

print(i)
```

<h1>Pass Statement with For Loop

```python
# An empty loop
for i in 'hieveryone':
    pass
print(i)
```

<h1>Else Statement with For Loops

```python
for i in range(1, 4):
    print(i)
else:  # Executed because no break in for
    print("No Break\n")
```

<h1>Using Enumerate with for loop

```python
li = ["eat", "sleep", "repeat"]

for i, j in enumerate(li):
    print (i, j)
```

<h1>Nested For Loops in Python

```python
for i in range(1, 4):
    for j in range(1, 4):
        print(i, j)
```

<h>Iterating by the Index of Sequences

```python
list = ["slow", "fast", "slow"]
for index in range(len(list)):
    print(list[index])
```
