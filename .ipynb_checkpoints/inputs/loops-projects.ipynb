<h1>Star patterns (*)</h1>

 <h1>1. Right-Angled Triangle</h1>

```python
n = 5
for i in range(1, n + 1):
    print('*' * i)
```

<h1>2. Left-Aligned Inverted Triangle</h1>

```python
n = 5
for i in range(n, 0, -1):
    print('*' * i)
```

 <h1>3. Right-Aligned Triangle</h1>

```python
n = 5
for i in range(1, n + 1):
    print(' ' * (n - i) + '*' * i)
```

<h1>4. Inverted Right-Aligned Triangle</h1>

```python
n = 5
for i in range(n, 0, -1):
    print(' ' * (n - i) + '*' * i)
```

 <h1>5. Pyramid</h1>

```python
n = 5
for i in range(1, n + 1):
    print(' ' * (n - i) + '*' * (2 * i - 1))
```

<h1>6. Inverted Pyramid</h1>

```python
n = 5
for i in range(n, 0, -1):
    print(' ' * (n - i) + '*' * (2 * i - 1))
```

<h1>7. Diamond</h1>

```python
n = 5
# Upper half
for i in range(1, n + 1):
    print(' ' * (n - i) + '*' * (2 * i - 1))
# Lower half
for i in range(n - 1, 0, -1):
    print(' ' * (n - i) + '*' * (2 * i - 1))
```

<h1>8. Hollow Square</h1>

```python
n = 5
for i in range(n):
    if i == 0 or i == n - 1:
        print('*' * n)
    else:
        print('*' + ' ' * (n - 2) + '*')
```

<h1>Sum of all even numbers from 1 to 100</h1>


```python
total = 0
for i in range(2, 101,2):
    total += i

print("Sum of even numbers from 1 to 100 is:", total)
```
