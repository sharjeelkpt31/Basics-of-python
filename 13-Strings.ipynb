<h1>Strings</h1>
Strings in python are surrounded by either single quotation marks, or double quotation marks.

'hello' is the same as "hello".

You can display a string literal with the print() function:




```python
print("hello")
print('hello')
```

<h1>Quotes Inside Quotes</h1>
You can use quotes inside a string, as long as they don't match the quotes surrounding the string:



```python
print("it's alright")
print("he is called'johnny'")
print('He is called "Johnny"')
```

<h1>Assign String to a Variable</h1>
Assigning a string to a variable is done with the variable name followed by an equal sign and the string:



```python
a="hello"
print(a)
```

<h1>Multiline Strings</h1>
You can assign a multiline string to a variable by using three quotes:



```python
a = """Lorem ipsum dolor sit amet,
consectetur adipiscing elit,
sed do eiusmod tempor incididunt
ut labore et dolore magna aliqua."""
print(a)
```

<h1>Or three single quotes:</h1>



```python
a = '''Lorem ipsum dolor sit amet,
consectetur adipiscing elit,
sed do eiusmod tempor incididunt
ut labore et dolore magna aliqua.'''
print(a)
```

<h1>Strings are Arrays</h1>
Like many other popular programming languages, strings in Python are arrays of bytes representing unicode characters.

However, Python does not have a character data type, a single character is simply a string with a length of 1.

Square brackets can be used to access elements of the string.



```python
a = "Hello, World!"
print(a[1])
```

<h1>Looping Through a String</h1>
Since strings are arrays, we can loop through the characters in a string, with a for loop.



```python
for x in "banana":
  print(x)
```

<h1>String Length</h1>
To get the length of a string, use the len() function.



```python
x="banana"
print(len(x))
```

<h1>Check String</h1>
To check if a certain phrase or character is present in a string, we can use the keyword in.



```python
txt = "The best things in life are free!"
print("free" in txt)
```

<h1>Use it in an if statement:</h1>



```python
txt = "The best things in life are free!"
if "free" in txt:
  print("Yes, 'free' is present.")
```

<h1>Check if NOT</h1>
To check if a certain phrase or character is NOT present in a string, we can use the keyword not in.



```python
txt = "The best things in life are free!"
print("expensive" not in txt)
```

<h1>Use it in an if statement:</h1>

```python
txt = "The best things in life are free!"
if "expensive" not in txt:
  print("No, 'expensive' is NOT present.")
```

<h1>Slicing</h1>
You can return a range of characters by using the slice syntax.

Specify the start index and the end index, separated by a colon, to return a part of the string.



```python
b = "Hello, World!"
print(b[2:5])
```

<h1>Slice From the Start</h1>
By leaving out the start index, the range will start at the first character:



```python
b = "Hello, World!"
print(b[:5])
```

<h1>Slice To the End</h1>
By leaving out the end index, the range will go to the end:



```python
b = "Hello, World!"
print(b[2:])
```

<h1>Negative Indexing</h1>
Use negative indexes to start the slice from the end of the string:


```python
b = "Hello, World!"
print(b[-5:-2])
```

<h1>Upper Case</h1>


```python
a = "Hello, World!"
print(a.upper())
```

<h1>Lower Case
</h1>

<h1>Remove Whitespace</h1>
Whitespace is the space before and/or after the actual text, and very often you want to remove this space.



```python
a = " Hello, World! "
print(a.strip()) # returns "Hello, World!"
```

<h1>Replace String</h1>


```python
a = "Hello, World!"
print(a.replace("H", "J"))
```

<h1>Split String</h1>
The split() method returns a list where the text between the specified separator becomes the list items.



```python
a = "Hello, World!"
print(a.split(",")) # returns ['Hello', ' World!']
```

<h1>String Concatenation</h1>
To concatenate, or combine, two strings you can use the + operator.



```python
a = "Hello"
b = "World"
c = a + b
print(c)
```

<h1>To add a space between them, add a " ":



```python
a = "Hello"
b = "World"
c = a + " " + b
print(c)
```

<h1>F-Strings</h1>
F-String was introduced in Python 3.6, and is now the preferred way of formatting strings.

To specify a string as an f-string, simply put an f in front of the string literal, and add curly brackets {} as placeholders for variables and other operations.



```python
age = 36
txt = f"My name is John, I am {age}"
print(txt)
```

<h1>Placeholders and Modifiers</h1>
A placeholder can contain variables, operations, functions, and modifiers to format the value.



```python
price = 59
txt = f"The price is {price} dollars"
print(txt)
```

<h1>Display the price with 2 decimals:



```python
price = 59
txt = f"The price is {price:.2f} dollars"
print(txt)
```

<h1>Perform a math operation in the placeholder, and return the result:</h1>



```python
txt = f"The price is {20 * 59} dollars"
print(txt)
```

<h1>Escape Character</h1>

To insert characters that are illegal in a string, use an escape character.

An escape character is a backslash \ followed by the character you want to insert.

An example of an illegal character is a double quote inside a string that is surrounded by double quotes:



```python
txt = "We are the so-called \"Vikings\" from the north."
txt
```

<h1>Python String Methods</h1>

<h1>Python String capitalize() Method

```python
txt = "hello, and welcome to my world."

x = txt.capitalize()

print (x)
```

<h1>Python String casefold() Method</h1>

```python
txt = "Hello, And Welcome To My World!"

x = txt.casefold()

print(x)
```

<h1>Python String center() Method</h1>

```python
txt = "banana"

x = txt.center(20)

print(x)
```

<h1>Python String count() Method



```python
txt = "I love apples, apple are my favorite fruit"

x = txt.count("apple")

print(x)
```

<h1>Python String endswith() Method



```python
txt = "Hello, welcome to my world."

x = txt.endswith(".")

print(x)
```

<h1>Python String expandtabs() Method



```python
txt = "H\te\tl\tl\to"

x =  txt.expandtabs(2)

print(x)
```

```python
txt = "Hello, welcome to my world."

x = txt.find("welcome")

print(x)
```
