<h1>File Handling</h1>


<p>File handling is an important part of any web application.<br>
Python has several functions for creating, reading, updating, and deleting files.</p>



<h1>Syntax</h1>
To open a file for reading it is enough to specify the name of the file:



```python
f = open("demofile.txt")
```

```python
f = open("demofile.txt", "rt")
```

<h>Because "r" for read, and "t" for text are the default values, you do not need to specify them.



```python

with open("demofile.txt", "w") as file:
    file.write("Hello! Welcome to demofile.txt\n")
    file.write("This file is for testing purposes.\n")
    file.write("Good Luck!\n")

print("✅ demofile.txt created successfully.")
```

<h1>Using the with statement</h1>
You can also use the with statement when opening a file:



```python
with open("demofile.txt") as f:
  print(f.read())
```

<h1>Close Files</h1>
<h>It is a good practice to always close the file when you are done with it.
If you are not using the with statement, you must write a close statement in order to close the file:</h>



```python
f = open("demofile.txt")
print(f.readline())
f.close()
```

```python
with open("demofile.txt") as f:
  print(f.read(6))
```

<h1>Read Lines</h1>
You can return one line by using the readline() method:



```python
with open("demofile.txt") as f:
  print(f.readline())
```

By looping through the lines of the file, you can read the whole file, line by line:



```python
with open("demofile.txt") as f:
  for x in f:
    print(x)
```

<h1>Write to an Existing File</h1>
To write to an existing file, you must add a parameter to the open() function:

"a" - Append - will append to the end of the file

"w" - Write - will overwrite any existing content



```python
with open("demofile.txt", "a") as f:
  f.write("Now the file has more content!")

#open and read the file after the appending:
with open("demofile.txt") as f:
  print(f.read())
```

<h1>Overwrite Existing Content</h1>
To overwrite the existing content to the file, use the w parameter:



```python
with open("demofile.txt", "w") as f:
  f.write("Woops! I have deleted the content!")

#open and read the file after the overwriting:
with open("demofile.txt") as f:
  print(f.read())
```

<h1>Create a New File</h1>
To create a new file in Python, use the open() method, with one of the following parameters:

"x" - Create - will create a file, returns an error if the file exists

"a" - Append - will create a file if the specified file does not exists

"w" - Write - will create a file if the specified file does not exists



```python
f = open("myfile.txt", "x")
```

<h1>Delete a File</h1>
To delete a file, you must import the OS module, and run its os.remove() function:



```python
import os
os.remove("demofile.txt")
```

<h1>Check if File exist:</h1>
To avoid getting an error, you might want to check if the file exists before you try to delete it:



```python
import os
if os.path.exists("demofile.txt"):
  os.remove("demofile.txt")
else:
  print("The file does not exist")
```

<h1>Delete Folder</h1>
To delete an entire folder, use the os.rmdir() method:



```python
import os
os.rmdir("myfolder")
```
